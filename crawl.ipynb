```python
import requests
from bs4 import BeautifulSoup
import re
import os
import pandas as pd

data = {}
for i in range(5169,5635):
  print(f'TRY TO INDEX: {i}')
  try:
    # 크롤링할 웹사이트 주소
    url = f'http://www.416memory.org/collection_col1/view/{i}'

    # 웹사이트에서 HTML 가져오기
    response = requests.get(url)
    html = response.text



    # BeautifulSoup 객체 생성
    soup = BeautifulSoup(html, 'html.parser')

    # 원하는 데이터 추출하기 (예: 모든 제목 태그)
    # print(soup.select_one('#content > div.mainArea > div > div > div.flbox > div > div.normal-txt > table > tbody > tr:nth-child(1) > td > img'))
    description = soup.find('td',class_='img_desc')
    description = description.get_text()
    contents = soup.select('.galleryview > ul li')
    name = ''
    aci = 65
    di = { "id" : i, "name" : name}
    
    for content in contents:
      filename = content.find('b').get_text()
      img = content.find('span', class_='img')
      style = img['style']
      if name == '' and  "사랑하는 " in filename:
        name = filename.replace("사랑하는 ",'')
        di["name"] =  name
        if not os.path.exists(f'./{i}'):
          os.mkdir(f'./{i}')
      url_match = re.search(r'url\((.*?)\)', style)
      if url_match:
        background_image_url = url_match.group(1)
        if "pdf" in background_image_url:
          background_image_url= background_image_url.replace("pdf",'jpg')
        else:
          background_image_url = background_image_url.replace(":S",'')
        full_image_url = "http://www.416memory.org" + background_image_url
      # Download and save the image
      try:
          response = requests.get(full_image_url)
          response.raise_for_status()  # Check if the request was successful
          pathname= os.path.basename(background_image_url)
          with open(f'./{i}/{filename}.jpg', 'wb') as file:
              file.write(response.content)

          print(f"Image successfully downloaded and saved")
      except requests.RequestException as e:
          pass
      except Exception as e:
          pass
      di[chr(aci)] = filename
      aci += 1
    
    data[i] = di
      
      # style= style.replace("pdf",'jpg')
      # style= style.replace("pdf",'jpg')
      # url_match = re.search(r'url\((.*?)\)', style)
      # if url_match:
      #   background_image_url = url_match.group(1)
      #   full_image_url = "http://www.416memory.org" + background_image_url
      # # Download and save the image
      # try:
      #     response = requests.get(full_image_url)
      #     response.raise_for_status()  # Check if the request was successful
      #     pathname= os.path.basename(background_image_url)
      #     with open(f'./{filename}.jpg', 'wb') as file:
      #         file.write(response.content)

      #     print(f"Image successfully downloaded and saved")
      # except requests.RequestException as e:
      #     pass
      # except Exception as e:
      #     pass
      # else:
      #   print("No background image URL found")
  except:
    continue


result = pd.DataFrame.from_dict(data,orient='index')
# result = result.transpose()
result.to_excel('./result.xlsx')
```
