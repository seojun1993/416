```python
from selenium import webdriver
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
browser = webdriver.Chrome()

try:
  url = 'https://www.goe416.go.kr/?p=27&page=3&viewType=class&searchTxt='
  browser.get(url)
  html = browser.page_source
  wait = WebDriverWait(browser,10)

  wait.until(EC.presence_of_element_located((By.CLASS_NAME,'list')))
  # 웹사이트에서 HTML 가져오기
  # response = requests.get(url)
  # html = response.text

  result = []
  # BeautifulSoup 객체 생성
  soup = BeautifulSoup(html, 'html.parser')

  swiper = soup.find('div',id='viewType_class')
  wrapper= swiper.find('div',class_='swiper-wrapper')
  
  li = wrapper.find_all('div',class_='swiper-slide')
  for i in li:
    className = i.find('h3',class_='std-slider-class').get_text()
    item = i.find('div',class_='list')
    stus  = item.find_all('button')
    for button in stus:
      match = re.search(r'idx=(\d+)', button['onclick'])
      
      if match:
        idx_number = match.group(1)
        name = button.get_text()
        result.append({"idx_number":idx_number,"name":name,"className":className, "href" :button['onclick'].replace("location.href=",'')})
      else:
        print("idx 숫자를 찾을 수 없습니다.")
      
except Exception as e:
  print(e)
finally:
  browser.quit()
  
df = pd.DataFrame(result)
df.head()

df.to_excel('./id.xlsx')
```
